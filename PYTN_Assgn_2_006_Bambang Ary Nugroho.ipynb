8. Menganalisis Confidence Intervals 

```python
df.head(3)
```

```python
dx = df.dropna(subset=["SALE PRICE", "BOROUGH"])

pd.crosstab(dx['SALE PRICE'], dx['BOROUGH'])
```

```python
p_fm = 1826/(1826+22)
p_fm
```

```python
n = 1826+22
n
```

```python
br_2 = np.sqrt(p_fm * (1 - p_fm) / n)
br_2
```

```python
z_score = 1.96
lcb = p_fm - z_score* br_2 #lower limit of the CI
ucb = p_fm + z_score* br_2 #upper limit of the CI
print('Confidence interval ',lcb, ucb)
```

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

##### Hipotesis Null: harga jual BOROUGH1>BOROUGH2 
##### Hipotesis Alternatif: harga jual BOROUGH1<BOROUGH2

```python
b1 = df[df['BOROUGH'] == 1]
b2 = df[df['BOROUGH'] == 2]


n1 = len(b1)
mu1 = b1["SALE PRICE"].mean()
sd1 = b1["SALE PRICE"].std()

n2 = len(b2)
mu2 = b2["SALE PRICE"].mean()
sd2 = b2["SALE PRICE"].std()

print(n1, mu1, sd1)
print(n2, mu2, sd2)
```

```python
sm.stats.ztest(b1["SALE PRICE"].dropna(), 
               b2["SALE PRICE"].dropna(),
               alternative='two-sided')
```
